```python
import numpy as np
import torch
from ml.scripts import make_mock_df
from ml.scripts.trainers import MLPTrainer
from worker.models.models import MLPRegressor
from worker.models.specs import ModelConfig
```

```python
df = make_mock_df(n_features=10, n_targets=1, n_samples=1000, random_state=42, noise=0.1)
df.head()
```

```python
feature_cols = [c for c in df.columns if c.startswith("x")]
target_cols = [c for c in df.columns if c.startswith("y")]

X = df[feature_cols].to_numpy()
y = df[target_cols].to_numpy()

split = int(0.8 * len(df))
train_x, test_x = X[:split], X[split:]
train_y, test_y = y[:split], y[split:]
```

```python
spec = ModelConfig(
    model_type="mlp_regressor",
    features=feature_cols,
    targets=target_cols,
    model_kwargs={"hidden_dims": [64, 64], "dropout": 0.0},
)
model = MLPRegressor(spec)
```

```python
trainer = MLPTrainer(model, lr=1e-3, weight_decay=0.0)
metrics = trainer.train(train_x, train_y, epochs=50, batch_size=64)
sum(metrics["losses"]) / len(metrics["losses"])
```

```python
preds = model.predict(test_x, device="cpu")
mse = np.mean((preds - test_y) ** 2)
mse
```
